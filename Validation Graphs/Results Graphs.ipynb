# Results

Loading important libraries and metric histories for both the models

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300

# train results

t5_metrics = pd.read_csv('./FLAN T5 Small Fine_Tuning on NvidiaQA/metrics_history_FLAN_T5_small.csv')
t5_metrics.index = range(1, len(t5_metrics) + 1)
# display(t5_metrics)

gpt2_metrics = pd.read_csv('./DistilGPT2 Fine_Tuning on NvidiaQA/metrics_history_GPT2.csv')
gpt2_metrics.index = range(1, len(gpt2_metrics) + 1)
# display(gpt2_metrics)


# test results

t5_test_metrics = pd.read_csv('./FLAN T5 Small Fine_Tuning on NvidiaQA/FlanT5_small_Test_Results.csv')
# display(t5_test_metrics)

gpt2_test_metrics = pd.read_csv('./DistilGPT2 Fine_Tuning on NvidiaQA/GPT2_Test_Results.csv')
# display(gpt2_test_metrics)
```

## Training Performance

### 1. F1, Precision, Recall

```python
# Plot Precision, Recall, and F1 Score in one plot
plt.figure(figsize=(14, 8))

# Plotting Precision
plt.plot(t5_metrics['precision'], label='FLAN-T5-small Precision', marker='o', linestyle=':', color='tab:red', linewidth=2)
plt.plot(gpt2_metrics['precision'], label='DisitlGPT2 Precision', marker='o', linestyle=':', color='tab:blue', linewidth=2)


# Plotting Recall
plt.plot(t5_metrics['recall'], label='FLAN-T5-small Recall', marker='s', linestyle='--', color='tab:red', linewidth=2)
plt.plot(gpt2_metrics['recall'], label='DisitlGPT2 Recall', marker='s', linestyle='--', color='tab:blue', linewidth=2)


# Plotting F1 Score
plt.plot(t5_metrics['f1'], label='FLAN-T5-small F1-Score', marker='^', linestyle='-', color='tab:red', linewidth=2)
plt.plot(gpt2_metrics['f1'], label='DisitlGPT2 F1-Score', marker='^', linestyle='-', color='tab:blue', linewidth=2)

# Adding plot title and labels
plt.title('Comparison of Precision, Recall, and F1 Scores over Epochs', fontsize=20)
plt.xlabel('Epochs', fontsize=16)
plt.ylabel('Scores', fontsize=16)
plt.legend(loc='upper left')
plt.grid(True)

plt.xticks(range(1, 6))

# Save the figure in high-definition
plt.savefig('./Graphs/f1_p_r_comparison_hd.png', dpi=300)

# Show the plot
plt.show()
```

### 6. Latency and Memory Usage

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate average values for latency and memory usage for both models
t5_avg_latency = t5_metrics['latency'].mean()
t5_avg_memory = t5_metrics['memory_usage'].mean()
gpt2_avg_latency = gpt2_metrics['latency'].mean()
gpt2_avg_memory = gpt2_metrics['memory_usage'].mean()

# Data to plot
categories = ['Latency (s)', 'Memory Usage (MB)']
flan_data = [t5_avg_latency, t5_avg_memory]
gpt_data = [gpt2_avg_latency, gpt2_avg_memory]

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting latency bars for both models
latency_bars_f = ax1.bar(x[0] - width/2, flan_data[0], width, label='FLAN-T5-small', color='tab:red')
latency_bars_g = ax1.bar(x[0] + width/2, gpt_data[0], width, label='DistilGPT2', color='tab:blue')

# Set labels for the primary y-axis
ax1.set_ylabel('Latency (s)', fontsize=16)
ax1.set_title('Comparison of Average Latency and Memory Usage per Epoch', fontsize=18)
ax1.set_xticks(x)
ax1.set_xticklabels(categories)
ax1.tick_params(axis='y')

# Create a twin Axes sharing the same x-axis for memory usage
ax2 = ax1.twinx()

# Plotting memory usage bars for both models
memory_bars_f = ax2.bar(x[1] - width/2, flan_data[1], width, color='tab:red')
memory_bars_g = ax2.bar(x[1] + width/2, gpt_data[1], width, color='tab:blue')

# Set labels for the secondary y-axis
ax2.set_ylabel('Memory Usage (MB)', fontsize=16)
ax2.tick_params(axis='y')

# Adding a function to label the bars
def autolabel(bars, axis):
    for bar in bars:
        height = bar.get_height()
        axis.annotate(f'{height:.2f}',
                      xy=(bar.get_x() + bar.get_width() / 2, height),
                      xytext=(0, 3),  # 3 points vertical offset
                      textcoords="offset points",
                      ha='center', va='bottom')

autolabel(latency_bars_f, ax1)
autolabel(latency_bars_g, ax1)
autolabel(memory_bars_f, ax2)
autolabel(memory_bars_g, ax2)

fig.tight_layout()
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, 1), ncol=2)

# Save the figure in high-definition
plt.savefig('./Graphs/latency_memory_comparison_hd.png', dpi=300)

# Display the plot
plt.show()

```
